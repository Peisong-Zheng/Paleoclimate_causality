```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import timedelta

# Create a sample DataFrame
data = {
    'Task': [
        'Learn Causal Inference Methods',
        'Investigate Last Deglaciation with GMM & Draft Paper',
        'Present at the First Year Forum',
        'Write a Review of Causal Models',
        'Prepare First Year Report',
        'Study DO Event Causality',
        'Write Paper on DO Event Causality',
        'Research Last Deglaciation Causality',
        'Write Paper on Last Termination Causality',
        'Examine Causality in the Carbon Cycle',
        'Write Paper on Carbon Cycle Causality',
        'Draft Thesis'
    ],
    'Start': [
        '2023-10-01', '2023-07-01', '2023-12-15', '2024-03-15', '2024-04-01',
        '2024-07-01', '2024-12-01','2024-11-01', '2025-04-01', '2025-06-01', '2025-11-01', '2026-01-01'
    ],
    'Finish': [
        '2024-03-01', '2024-06-01', '2024-02-01', '2024-05-01', '2024-06-10',
        '2025-02-01', '2025-03-01','2025-06-01', '2025-07-01', '2026-01-01', '2026-02-01', '2026-07-01'
    ],
}

df = pd.DataFrame(data)
df['Start'] = pd.to_datetime(df['Start'])
df['Finish'] = pd.to_datetime(df['Finish'])

# Plotting
fig, ax = plt.subplots(figsize=(10, 4),dpi=600)
for i, task in df.iterrows():
    ax.barh(task['Task'], task['Finish'] - task['Start'], left=task['Start'], color='k',alpha=0.4)

# Formatting the chart
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))  # Set to every month
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
# reverse the y-axis
plt.gca().invert_yaxis()
plt.xticks(rotation=45)
plt.xlabel('Date')
# plt.ylabel('Task')
# plt.title('Gantt Chart')
# plt.grid(True)

# Turn off the box
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Show plot
plt.tight_layout()
plt.show()
```
